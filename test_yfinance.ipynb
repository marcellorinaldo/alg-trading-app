```python
import plotly.graph_objects as go
import matplotlib.pyplot as plt
import yfinance as yf

# auxiliary functions
def plot_candlestick(times, open_prices, high_prices, low_prices, close_prices):
    """
    Candlestick barplot (refer to https://plotly.com/python/candlestick-charts/ for documentation).
    """
    fig = go.Figure(data=[go.Candlestick(
        x=times,
        open=open_prices,
        high=high_prices,
        low=low_prices,
        close=close_prices)])

    fig.show()
```

```python
data = yf.download(tickers='AAPL', period='1d', interval='5m')
```

```python
data
```

```python
plot_candlestick(data.index, data['Open'], data['High'], data['Low'], data['Adj Close'])
```
